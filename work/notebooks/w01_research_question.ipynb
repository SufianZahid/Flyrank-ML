## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Load the data directly using the raw GitHub URL to avoid path issues in Colab
url = "https://raw.githubusercontent.com/SufianZahid/Flyrank-ML/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Metric 1: Total volume of pages to prioritize
total_pages = len(df)
print(f"Total Content Assets in Starter Slice: {total_pages}")

# Metric 2: Baseline proportion of actively declining pages
trend_counts = df['trend_direction'].value_counts()
trend_pct = df['trend_direction'].value_counts(normalize=True) * 100
print(f"\nTrend Distribution:\n{trend_counts}")
print(f"Percentage of Actively Declining Pages: {trend_pct.get('down', 0):.2f}%")

# Metric 3: High-Exposure Opportunity Windows (Pages with significant audience reach)
high_exposure = len(df[df['impressions_90d'] >= 500])
print(f"\nPages with Significant Search Volume (>= 500 impressions): {high_exposure}")
```
